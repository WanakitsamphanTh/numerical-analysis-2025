```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
errors = pd.read_csv("./errors-1.6.csv",delimiter='\t',header=None,names = ["i","float","double"])
print(errors)
```

```python
fig, ax = plt.subplots()
ax.set_xlim(10,10000)
ax.set_xscale('log',base=10)
ax.set_yscale('log',base=10)
ax.set_xlabel('Number of Iterations')
ax.set_ylabel('Absolute Error')
ax.plot(errors['i'],(errors["float"]),linestyle='solid',color='black')
ax.plot(errors['i'],(errors["double"]),linestyle='dotted', color='black')
ax.legend(['float','double'])
```
