```python
import numpy as np
from functools import reduce

def daikei(N):
    h = 1 / N
    S = 0
    for i in range(N):
        xs = [1 + i * h, 1 + (i+1) * h]
        S = S + h / 2 * (1 / xs[0] + 1 / xs[1])
    return S

def simpson(N):
    h = 1 / N
    S = 0
    for i in range(1,N,2):
        xs = [1 + (i-1) * h, 1 + i * h, 1 + (i+1) * h]
        S = S + h / 3 * (1 / xs[0] + 4 * 1 / xs[1] + 1 / xs[2])
    return S

y = np.log(2)

print("h\t\t\tdaikei\t\t\tsimpson")
print("%.10f\t\t%.10f\t\t%.10f"%(h,y-daikei(N),y-simpson(N)))
```
