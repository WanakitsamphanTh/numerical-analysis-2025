```python
import numpy as np
x = 1
h = 0.1
forward = np.cos(x) - (np.sin(x + h) - np.sin(x)) / 0.1
backward = np.cos(x) - (np.sin(x) - np.sin(x - h)) / 0.1
center = np.cos(x) - (np.sin(x + h) - np.sin(x - h)) / (2 * 0.1)

print("h\t\t\tforward\t\t\tbackward\t\tcenter")
print("%.8f\t\t%.8f\t\t%.8f\t\t%.8f"%(h,forward,backward,center))
```
