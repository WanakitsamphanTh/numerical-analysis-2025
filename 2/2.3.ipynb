```python
import numpy as np
x = 1
H = [1/(10**(x/10)) for x in range(1,150)]
forward = [np.abs(1/x - (np.log(x+h) - np.log(x)) / h) for h in H]
backward = [np.abs(1/x - (np.log(x) - np.log(x-h)) / h) for h in H]
center = [np.abs(1/x - (np.log(x+h) - np.log(x-h)) / (2*h)) for h in H]
```

```python
print(H)
print(forward)
print(backward)
print(center)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.set_xscale('log',base=10)
ax.set_yscale('log',base=10)
ax.plot(H,forward,linestyle='solid',color='black')
ax.plot(H,backward,linestyle='solid',color='black')
ax.plot(H,center,linestyle='dotted',color='gray')
```
